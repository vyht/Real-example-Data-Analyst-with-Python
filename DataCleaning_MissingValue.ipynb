##CLEANING DATA WITH DATAFRAME

```python
df = pd.DataFrame({
    'Column A': [1, np.nan, 30, np.nan],
    'Column B': [2, 8, 31, np.nan],
    'Column C': [np.nan, 9, 32, 100],
    'Column D': [5, 8, 34, 110],
})
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
x = df.dropna() # xa ton b ct/dng c cha gi tr null
x
```

```python
y = df.dropna(axis='columns') # xa ct c cha null, hoc axis = 1
y
```

```python
z = df.dropna(axis='rows') # xa hng c cha null, hoc axis=0
z
```

```python
m = df.dropna(how='all') #gi nguyn
m
```

```python
n = df.dropna(how='any') #xa ht nhng ct v dng c cha gi tr null (default)
n
```

```python
o = df.dropna(thresh=3) # drop ht gi tr theo dng (ct gi nguyn s lng) nhng  li 3 hng u tin (0,1,2) cho d c null hay ko
o
```

```python
p = df.dropna(thresh = 3, axis=1)  # drop ht gi tr theo ct (hng gi nguyn s lng) nhng  li 3 ct u tin (0,1,2) cho d c null hay ko
p
```

##FILLING NULL VALUES

```python
s
```

```python
s.fillna(0) # fill tt c gi tr null bng 0
```

```python
s.fillna(s.mean()) # fill tt c null bng mean
```

```python
# fill bng gi tr gn null pha trc: foward fill
s.fillna(method='ffill')
```

```python
# fill bng gi tr gn null pha sau: back fill
s.fillna(method = 'bfill')
```

##FILLING NULL VALUES WITH DATAFRAME

```python
df
```

```python
df.fillna({'Column A':0, 'Column B':99, 'Column C':df['Column C'].mean()}) # fill ct A = 0, B = 99, C = mean
```

```python
df.fillna(method='ffill', axis=0)
```

```python
df.fillna(method='ffill', axis=1)
```

```python
df.fillna(method='bfill', axis=0)
```

```python
df.fillna(method='bfill', axis=1)
```
