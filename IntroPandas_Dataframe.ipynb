##INDEXING, SELECTION AND SLICING

```python
df.iloc[-1] #iloc truy cập theo vị trí (position), không dựa trên tên mới đặt, chỉ cần là ở vị trí đó thì sẽ ra
```

##CONDITIONAL SELECTION (BOOLEAN ARRAYS)

```python
df['Population']>70
```

```python
df[df['Population']>70] # không cần loc thì sẽ lấy hết các cột
```

```python
df.loc[df['Population']>70, 'Population'] # khi có loc sẽ lấy được cột muốn lấy
```

##DROPPING

```python
df.drop('Canada') # drop hàng hoặc df.drop('Canada', axis = 0) hoặc axis = rows
```

```python
df.drop(columns = 'GDP') # drop cột Hoặc nếu không dùng columns có thể dùng df.drop('GDP', axis = 1) hoặc axis = columns
```

##MODIFYING DATAFRAME

Adding a new column

```python
langs = pd.Series(['Canada', 'Germany','Italy']
                  , index = ['Canada', 'Germany','Italy']
                  , name = 'Language')
langs # tạo series mới cho language
```

```python
df['Language'] = langs # thêm cột
df
```


Replacing values per column

```python
df['Language'] = 'English' # thay thế tất cả giá trị trong cột Language bằng English
df
```

```python
df.rename(columns = {'HDI':'Human Development Index', 'GDP':'Gross Domestic Product'}
          , index = {'United Kingdom':'UK', 'United States':'USA'})
df
```

Adding values

```python
df = pd.concat([df, pd.Series({'Population':3, 'GDP':5}, name = 'China').to_frame().T])
df
```

##CREATING NEW COLUMN

```python
df['GDP Per Capita'] = df['GDP'] / df['Population']
df
```
