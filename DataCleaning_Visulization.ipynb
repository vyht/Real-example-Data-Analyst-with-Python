#API - application programming interface GLOBAL

```python
x = np.arange(-10,11) # tạo ra mảng
x
```

```python
plt.figure(figsize=(10,5)) # chỉnh size
plt.title('My Graph') # tên biểu đồ
plt.plot(x, x**2) # x bình phương
plt.plot(x, (x**2)*-1) # -x bình phương
```

```python
plt.figure(figsize=(15,7))
plt.title('2 graph')

plt.subplot(1,2,1) # 1 hàng 2 cột và vẽ ở mảng 1
plt.plot(x, x**2)
plt.plot([0,0,0],[-10,0,100]) # plt.plot([0,0,0] - độ dài trục x chạy qua,[-10,0,100] - độ dài trục y chạy qua)

plt.subplot(1,2,2) # vẽ ở mảng 2
plt.plot(x,x**2)
plt.plot([-10,0,10],[20,20,20]) # plt.plot([-10,0,10] - độ dài trục x chạy qua,[20,20,20] - độ dài trục y chạy qua)
```

```python
fig, chart = plt.subplots(figsize=(12, 6)) # fig là phần khung, chart là phần để vẽ, cỡ 12,6, khi vẽ biểu đồ sẽ vẽ trong vùng 'chart'
# khi không sử dụng fig, chart... mà dùng plot.... thì maplotlib tự mặc định tạo ra khung và phần vẽ nhưng nếu muốn chỉnh phải chỉnh từng chi tiết
chart.plot(x,x**2, color = 'red', marker = '*',linestyle = '-', markersize = 10, label= 'X^2')
chart.plot(x,-1*(x**2), color = 'green', linestyle = '--', label = '-x^2')
chart.legend() # để hiện ra label
chart.set_xlabel('x')
chart.set_ylabel('x^2')
```

## The subplot2grid command
plt.subplot2grid (shape, loc, rowpan, colpan)

*   kích thước lưới: (hàng, cột)
*   vị trí bắt đầu: (hàng, cột)
*   rowpan: số hàng chiếm
*   colpan: số cột chiếm



```python
plt.figure(figsize=(14, 6))

af1 = plt.subplot2grid((3,3), (0,0), colspan=3) # cỡ 3-3, vị trí bắt đầu hàng 0, cột 0, chiếm 3 cột
af2 = plt.subplot2grid((3,3), (1,0), colspan=2) # cỡ 3-3, vị trí bắt đầu hàng 1, cột 0, chiếm 2 cột
af3 = plt.subplot2grid((3,3), (1,2), rowspan=2) # cỡ 3-3, vị trí bắt đầu hàng 1, cột 2, chiếm 2 hàng
af4 = plt.subplot2grid((3,3), (2,0))
af5 = plt.subplot2grid((3,3), (2,1))
```

##Scatter plot

*   fig = plt.figure(figsize=()) kết quả chỉ trả về mỗi figure còn axes sẽ tự thêm, điều này giúp linh hoạt khi muốn tự điều chỉnh axes
*   plt.subplots(figsize = ()) kết quả trả về có figure và axes, sử dụng khi chỉ cần 1 mảng visualization

```python
N = 100
x = np.random.randn(N) # random giá trị x trong khoảng từ 0-50
y = np.random.randn(N) # random giá trị y trong khoảng từ 0-50
colors = np.random.randn(N) # random màu trong khoảng từ 0-50
area = np.pi*(20 * np.random.rand(N))**2 # bán kính vòng tròn
```

```python
plt.figure(figsize = (12,6))
plt.scatter(x,y, s = area, c = colors, alpha = 0.3, cmap = 'viridis', marker='o', edgecolors = 'black', linewidths = 1)
plt.colorbar() # show thanh màu
plt.show()
```

```python
fig = plt.figure(figsize=(13,6))
ax1 = fig.add_subplot(1,2,1) # chia khung trong fig
plt.scatter(x,y, s = area, c = colors, alpha = 0.5, cmap = 'Pastel1')
plt.colorbar()

ax2 = fig.add_subplot(1,2,2)
plt.scatter(x,y, s = area, c = colors, alpha = 0.6, cmap = 'Pastel2', marker = '*')
plt.colorbar()
```

##Histogram

```python
values = np.random.randn(2000)
plt.subplots(figsize = (12,6))
plt.hist(values, color='green', alpha = 0.3, bins=100, histtype = 'bar', edgecolor='steelblue')
plt.xlim(-5,5) # x chạy từ -5 đến 5
plt.show()
```


## KDE (kernel density estimation)

```python
from scipy import stats

density = stats.kde.gaussian_kde(values)
density
```

```python
plt.subplots(figsize=(12, 6))

values2 = np.linspace(min(values)-10, max(values)+10, 100)

plt.plot(values2, density(values2), color='black') # màu viền
plt.fill_between(values2, 0, density(values2), alpha=0.3, color='steelblue') # màu fill bên trong kde
plt.xlim(-5,5)

plt.show()
```

##Combine plot

```python
plt.subplots(figsize=(12, 6))
plt.hist(values, bins=100, alpha=0.4, density=1, histtype='bar', color='steelblue', edgecolor='green')
plt.plot(values2, density(values2), color='red', linewidth=2.0)
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

## Bar plots

```python
Y = np.random.rand(1, 5)[0]
Y2 = np.random.rand(1, 5)[0]
```

```python
plt.figure(figsize=(12, 4))
barWidth = 0.4 # độ lớn bar
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='teal', alpha = 0.7)
plt.show()
```

```python
plt.figure(figsize=(12, 4))
barWidth = 0.4
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='teal', label='Label Y', alpha = 0.7)
plt.bar(np.arange(len(Y2)), Y2, width=barWidth, color='burlywood', bottom=Y, label='Label Y2', alpha = 0.7)

plt.legend()
plt.show()
```
