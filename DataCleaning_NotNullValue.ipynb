```python
df = pd.DataFrame({
    'Sex': ['M', 'F', 'F', 'D', '?'],
    'Age': [29, 30, 24, 290, 25],
})
df
```

##FIND UNIQUE VALUES

```python
df['Sex'].unique() # tìm ra các giá trị có xuất hiện trong cột Sex
```

```python
df['Sex'].value_counts()
```

```python
df['Sex'].replace('D','F') #đ đổi D thành F
```

Có thể nhầm lẫn trong quá trình nhập liệu nên F bị nhầm thành D và có thể M bị nhầm thành N, nhưng nếu làm việc với nhiều cột thì cần sử dụng tới level "DataFrame"

```python
df.replace([{'Sex':{'N':'M','D':'F'}}, {'Age':{290:29}}])
```

```python
df[df['Age']>100]
```

##DUPLICATE

```python
ambassadors = pd.Series([
    'France',
    'United Kingdom',
    'United Kingdom',
    'Italy',
    'Germany',
    'Germany',
    'Germany',
], index=[
    'Gérard Araud',
    'Kim Darroch',
    'Peter Westmacott',
    'Armando Varricchio',
    'Peter Wittig',
    'Peter Ammon',
    'Klaus Scharioth '
])
ambassadors
```

```python
ambassadors.duplicated()
```

```python
ambassadors.duplicated(keep='last') # giữ lại giá trị cuối, nghĩa là giá trị xuất hiện cuối cùng sẽ không bị tính là duplicated
```

```python
ambassadors.duplicated(keep = 'first') # giữ lại giá trị đầu tiên, nghĩa là giá trị xuất hiện đầu tiên sẽ không bị tính là duplicated
```

```python
ambassadors.drop_duplicates() #drop tất cả những giá trị bị tính là duplicated
```

```python
ambassadors.drop_duplicates(keep = 'last') # giá trị cuối trong các giá trị duplicated sẽ được giữ lại
```

```python
ambassadors.drop_duplicates(keep= 'first') # giá trị đầu trong các giá trị duplicated sẽ được giữ lại
```

```python
ambassadors.drop_duplicates(keep = False) # những giá trị không xuất hiện duplicated
```

##DUPLICATED IN DATAFRAME

```python
players = pd.DataFrame({
    'Name': [
        'Kobe Bryant',
        'LeBron James',
        'Kobe Bryant',
        'Carmelo Anthony',
        'Kobe Bryant',
    ],
    'Pos': [
        'SG',
        'SF',
        'SG',
        'SF',
        'SF'
    ]
})
players
```

```python
players.duplicated()
```

```python
players.duplicated(subset='Name') #duplicated theo cột Name
```

```python
players.duplicated(subset='Pos') #duplicated theo cột Pos
```

```python
players.drop_duplicates()
```

```python
players.drop_duplicates(subset='Name', keep='first')
```

##TEXT HANDLING
Muốn áp dụng các phương thức xử lý chuỗi thì phải thêm .str vì dạng đang xử lý là str



```python
data = pd.DataFrame({
    'Data': [
        '1987_M_US _1',
        '1990?_M_UK_1',
        '1992_F_US_2',
        '1970?_M_   IT_1',
        '1985_F_I  T_2'
]})
data
```

Những thành phần như năm sinh, giới tính, quốc gia và số con trong gia đình đáng lẽ phải ở cột khác nhau nhưng chung 1 cột và cách nhau bởi dấu gạch dưới => trong python có thể tách những thành phần này ra bằng split()

```python
data['Data'].str.split('_') #tách
```

```python
df = data['Data'].str.split('_', expand = True) # chuyển thành bảng
df
```

```python
df.columns = ['Year', 'Gender', 'Country', 'NoChild']
df
```

```python
df['Year'].str.contains('\?') #kiểm tra có dấu ? trong cột Gender bằng contains
```

```python
df['Country'].str.contains('U')
```

```python
df['Country'].replace(' ','') #xóa khoảng trắng
```
