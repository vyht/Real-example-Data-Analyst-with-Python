##**Dataset: market-price**


1.   header = None => không lấy tên cột
2.   na_values = [] => gán NaN cho những giá trị có chứa []
3.   names = [] => đổi tên cột
4.   dtype = {'':''} => đổi kiểu dữ liệu
5.   pd.to_datetime() => đổi kiểu dữ liệu datetime
6.   parse_dates=[0] => tự động đổi dữ liệu object sang datetime mà ko cần to_datetime, tham số 0 sẽ ko tự truyền hoặc tự động thay đổi, 0 là index của cột
7.   index_col =[] => chuyển cột thành index (nếu cột được chỉ định làm index là datetime thì khi thành index sẽ chuyển thành object





```python
df1 = pd.read_csv('/content/drive/MyDrive/for jobs/GG Colab/Read data/btc-market-price.csv')
df1
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/for jobs/GG Colab/Read data/btc-market-price.csv',
                  header=None,
                  na_values=['','-','?'],
                  names = ['Timestamp','Price'],
                  dtype={'Price':'float'})
df1.head()
```

##**Dataset: exam_review**



*   Trong trường hợp csv là file được phân cách bởi dấu , thì pd sẽ tự động chuyển đổi nhưng có một số file được phân cách bởi ;, >, ?... thì phải tự phân tách
1.   sep = '>' => phân tách

*   Số thập phân và đơn vị hàng ngàn có thể khác nhau ở dataset khác nhau, nếu cột có dấu , để ngăn cách số thập phân và đơn vị hàng ngàn thì có thể bị hiểu là object mà ko phải số
2.   decimal =',' hoặc thousands =','
3.   skiprows = 3 => bỏ qua 3 cột đầu tiên
4.   skip_blank_lines=False => ko bỏ qua những cột NaN, nếu True thì bỏ qua
5.   usecols = [1,2,3] => chỉ định những cột được sử dụng
*   Mục danh sách
