```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/freeCodeCamp/boilerplate-demographic-data-analyzer/main/adult.data.csv')
df.head()
```

```python
# Có bao nhiêu người thuộc mỗi chủng tộc
race_count = df['race'].value_counts()
race_count
```

```python
# Tuổi trung bình nam giới
dfmale = df.loc[df['sex']=='Male']
avgage = dfmale['age'].mean()
print(f'age avg of male if',avgage)
```

```python
# Tỷ lệ người có bằng tốt nghiệp
df_ed_bach = df.loc[df['education']=='Bachelors']
percentage_bachelors = df_ed_bach.shape[0]/df.shape[0]*100
print(f'{percentage_bachelors:.2f}%')
```

```python
# Tỷ lệ người có học vấn cao/thấp mà có thu nhập trên 50000
df['education'].value_counts() #high edu: Bachelors, Masters, Doctorate
```

```python
# tỷ lệ học vấn cao kiếm trên 50k
dfhigh = df.loc[(df['education']=='Bachelors') | (df['education']=='Masters') | (df['education']=='Doctorate') & (df['salary']=='>50K')]
higher_education = dfhigh.shape[0]/df.shape[0]*100
print(f'{higher_education:.2f}%')
```

```python
# tỷ lệ học vấn thấp kiếm trên 50k
dflow = df.loc[(df['education']!='Bachelors') & (df['education']!='Masters') & (df['education']!='Doctorate') & (df['salary']=='>50K')]
lower_education = dflow.shape[0]/df.shape[0]*100
print(f'{lower_education:.2f}%')
```

```python
# tỷ lệ học vấn cao
dfedhigh = df.loc[(df['education']=='Bachelors') | (df['education']=='Masters') | (df['education']=='Doctorate')]
high_education = dfedhigh.shape[0]/df.shape[0]*100
print(f'{high_education:.2f}%')
# tỷ lệ học vấn thấp
dfedlow = df.loc[(df['education']!='Bachelors') & (df['education']!='Masters') & (df['education']!='Doctorate')]
low_education = dfedlow.shape[0]/df.shape[0]*100
print(f'{low_education:.2f}%')
```

```python
# số giờ làm việc tối thiểu của 1 người mỗi tuần
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

```python
# Tỷ lệ phần trăm những người làm việc tối thiểu số giờ mỗi tuần có mức lương trên 50k đô la là bao nhiêu?
dfmin50 = df.loc[(df['hours-per-week']==min_work_hours)&(df['salary']=='>50K')]
num_min_workers = dfmin50.shape[0]
rich_percentage = dfmin50.shape[0]/df.shape[0]*100
print(num_min_workers)
print(f'{rich_percentage:.2f}%')
```

```python
# Quốc gia nào có tỷ lệ người dân có thu nhập trên 50.000 đô la cao nhất và tỷ lệ đó là bao nhiêu?
countover50 = df[df['salary']=='>50K'].groupby(df['native-country']).size()
countallover50 = df.groupby(df['native-country']).size()
rateover50 = countover50/countallover50*100
rateover50.fillna(0)
```

```python
highest_earning_country = rateover50.idxmax()
highest_earning_country_percentage = rateover50.max()

print(f"The country with the highest percentage of people earning over 50K is: {highest_earning_country}")
print(f"The percentage is: {highest_earning_country_percentage:.2f}%")
```

```python
# Xác định nghề nghiệp phổ biến nhất đối với những người có thu nhập trên 50.000 đô la ở Ấn Độ.
dfindia = df.loc[df['native-country']=='India']
count = dfindia[dfindia['salary']=='>50K'].groupby(dfindia['occupation']).size()
pop_ca = count.idxmax()
print(pop_ca)
```
